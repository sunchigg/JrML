```python
import matplotlib.pyplot as plt
import base64
import httplib2
import os
from googleapiclient import discovery
from oauth2client.client import GoogleCredentials
import json
import pandas as pd
```

```python
#請修改金鑰為自己的.json
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = '我是金鑰.json'
DISCOVERY_URL='https://{api}.googleapis.com/$discovery/rest?version={apiVersion}'
credentials = GoogleCredentials.get_application_default()
service = discovery.build('vision', 'v1', credentials=credentials,
                          discoveryServiceUrl=DISCOVERY_URL)
```

```python
def get_labels(photo_file):
    with open(photo_file, 'rb') as image:
        image_content = base64.b64encode(image.read())
        service_request = service.images().annotate(body={
            'requests': [{
                'image': {
                    'content': image_content.decode('UTF-8')
                },
                'features': [{
                    'type': 'LABEL_DETECTION',
                    'maxResults': 20
                }]
            }]
        })
        response = service_request.execute()
        return response['responses'][0]['labelAnnotations']
```

```python
#網路上的範例，不需理會。https://github.com/enakai00/jupyter_gcp/blob/master/Vision%20API%20Quick%20Tour.ipynb
labels = get_labels('/Users/francislin/Desktop/Loan_FM/EDM_PHOTO/K01.jpg')
for label in labels:
    print ('%s : %2d%%' % (label['description'], label['score']*100))
```

```python
#一個圖組有多張照片，透過for loop將全部的資訊辨識一遍
append_df = pd.DataFrame(columns=['description', 'score'])
for i in range(1,7):
    pre_label=get_labels("/Users/francislin/Desktop/Loan_FM/EDM_PHOTO/K0%i.jpg"%(i))
    aa= json.dumps(pre_label)
    df= pd.read_json(aa)
    pre_trans = (df.drop(['mid','topicality'], axis=1))
    append_df = append_df.append(pre_trans, ignore_index=True)
```

```python
#接續合併結果，將相同的label annotation取平均
label= (append_df.groupby(by='description', as_index=True).agg({'score': pd.Series.mean})).sort_values(by='score',ascending=False)
Set_K= label.transpose()
Set_K['Photo_key']='K'
Set_K.columns = Set_K.columns.str.replace(' ','_')
Set_K
```
