```python
with open('all_data.pickle','rb') as f:
    df= pickle.load(f)
```

```python
df.head()
```

```python
data_key=df[['id','cuisine']]
data = df.drop(['id','cuisine'],axis=1)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.5, svd_solver='full')
pca_data=pca.fit(data)
```

```python
print(pca.n_components_)
```

```python
principalComponents = pca.fit_transform(data)
principalDf = pd.DataFrame(data = principalComponents, columns = ['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10', 'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20', 'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'V29', 'V30', 'V31', 'V32', 'V33'])
```

```python
principalDf.head()
```

```python
ALL = pd.concat([data_key, principalDf], axis=1)
```
