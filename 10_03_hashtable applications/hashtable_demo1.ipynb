### Hash Table概念應用 O()<O(N+N)

```python
data=pd.read_csv('for_same_amt_class1.csv',encoding='big5')
```

```python
bad_txn_amt_time = {}
for i in range(0, len(data)):
    mount = data.loc[i,'Amount']
    if bad_txn_amt_time.get(mount) == None:
        bad_txn_amt_time[mount] = [data.loc[i,'Time']]
    else:
        bad_txn_amt_time[mount].append(data.loc[i,'Time'])
```

```python
#匯入全部的交易資料，含training/testing data
#共284807筆
data_all=pd.read_csv('for_same_amt.csv',encoding='big5')
```

```python
import time
t0=time.time() #計時

for i in range(0, len(data_all)):
    time_list = bad_txn_amt_time.get(data_all.loc[i,'Amount'])
    data_all.loc[i,'Same_amt_in_30min'] = 0 #創欄位，先設定30分鐘內同金額交易次數為0
    if type(time_list)==list: #如果該金額的時間list只有一筆就不是list，那也不必去滾有沒有相同時間的其他筆
        for t in time_list:
            if abs(t - data_all.loc[i,'Time']) < 1800:
                data_all.loc[i,'Same_amt_in_30min'] += 1
t1=time.time()
print (t1-t0)
```
